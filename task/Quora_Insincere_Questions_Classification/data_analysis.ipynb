# 导入数据

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
train_df = pd.read_csv('../input/quora-insincere-questions-classification/train.csv')
test_df = pd.read_csv('../input/quora-insincere-questions-classification/test.csv')
```

```python
train_df.head()
```

```python
test_df.head()
```

# 文本分析

```python
train_df['text_len'] = train_df['question_text'].apply(lambda x: len(x.split(' ')))
print(train_df['text_len'].describe())
```

```python
_ = plt.hist(train_df['text_len'], bins=200)
plt.xlabel('Text char count')
plt.title("Histogram of char count")
```

```python
np.percentile(train_df['text_len'].tolist(), 99)
```

```python
test_df['text_len'] = test_df['question_text'].apply(lambda x: len(x.split(' ')))
print(test_df['text_len'].describe())
```

```python
_ = plt.hist(test_df['text_len'], bins=200)
plt.xlabel('Text char count')
plt.title("Histogram of char count")
```

```python
np.percentile(test_df['text_len'].tolist(), 99)
```
